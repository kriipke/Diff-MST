```python
# We are now going to create a new directory and pick just the info we want from the yaml files and save in yaml files in the new directory
dir = os.path.join("/homes/ssv02/style_transfer/mix_style_transfer/mst/dataloaders")
metadata_dir = os.path.join(dir, "medley_metadata")
new_meta_dir = os.path.join(dir, "new_metadata")
if not os.path.exists(new_meta_dir):
    os.mkdir(new_meta_dir)
    print("made directory")
os.chdir(new_meta_dir)
print(os.getcwd())
```

```python
import os
import glob
metadata_file = glob.glob(os.path.join(metadata_dir, '*.yaml'))
print(len(metadata_file))
```

```python
from yaml import load, dump
from yaml import Loader, Dumper

dump_dir = new_meta_dir
for files in metadata_file:
    print(os.path.basename(files))
    
    dump_file = os.path.join(dump_dir, os.path.basename(files))
    with open(files, 'r') as f:
        data = load(f, Loader=Loader)
        #print(data)
        stems = {}
        for k, v in data.items():
            if k == 'stems':
                stems = v
        #print("stems derived")        
                #print(stems)
    
        #raw_track = 0
        raw = {}
        for stem_name, raw_tracks in stems.items():
            #print(stem_name)
            for raw_dict, stem_deets in raw_tracks.items():
                if raw_dict == 'raw':
                    for raw_name, raw_deets in stem_deets.items():
                        #print(raw_name)
                        #raw_track +=1
                        raw[raw_deets['filename']] = raw_deets['instrument']
        
        #print(raw_track)
        #print(len(raw.keys()))
        #print(raw)
        
    with open (dump_file, 'w') as f:
        dump(raw, f, Dumper=Dumper)
        print("dumped")
    

with open (dump_file, 'r') as f:
    data = load(f, Loader=Loader)

print(len(glob.glob(os.path.join(dump_dir, "*.yaml"))))
   
```
