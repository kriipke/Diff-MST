```python
import os 
import sys
import glob
from yaml import load, Loader, dump, Dumper
```

```python
metadata_dir = "/homes/ssv02/Diff-MST/medley_metadata"
mix_dirs = glob.glob(os.path.join(metadata_dir, "*"))
print(len(mix_dirs))
```

```python
unique_instruments = []
for mix_dir in mix_dirs:
    with open(mix_dir, 'r') as f:
            mdata = load(f, Loader=Loader)
            print("\n", os.path.basename(mix_dir))
            print(mdata.values())
            for value in mdata.values():
                if value not in unique_instruments:
                    unique_instruments.append(value)
    f.close()
print(len(unique_instruments)) 
print(unique_instruments)         
                
```

```python
instrument = {}
for mix_dir in mix_dirs:
    with open(mix_dir, 'r') as f:
            mdata = load(f, Loader=Loader)
            for values in mdata.values():
                if values in instrument:
                    instrument[values] += 1
                else:
                    instrument[values] = 1
    f.close()
dict(sorted(instrument.items()))
print(instrument)
                 
                
```
