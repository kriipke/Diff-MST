```python
import sys
import glob
import os
rootdir = ["/data/scratch/acw639/Medley/V1",
"/data/scratch/acw639/Medley/MedleyDB_V2/V2"]
dirs = []
for dir in rootdir:
    mix_dirs = glob.glob(os.path.join(dir, "*"))
    # remove items that are not directories
    mix_dirs = [mix_dir for mix_dir in mix_dirs if os.path.isdir(mix_dir)]
    mix_dirs = sorted(mix_dirs)  # sort
    print(len(mix_dirs))
    for mix_dir in mix_dirs:
        dirs.append(mix_dir)
    
print(len(dirs))
print(dirs[0])

```

```python
from collections import Counter
counter = {}
for dir in dirs:
    
    mix_id = os.path.basename(dir)
    track_filepaths = glob.glob(os.path.join(dir, f"{mix_id}_RAW", "*.wav"))
    num_tracks = len(track_filepaths)
    counter[mix_id] = num_tracks

    
    

```

```python
num_tracks_list = [i for i in counter.values()]


tracks_count = Counter(num_tracks_list)

print(tracks_count)
```
