```python
import glob
import sys
import os
import yaml
from yaml import load, dump
from yaml import Loader, Dumper
```

```python
import numpy as np
#load medleydb v1 and v2
#paste the directories that contain V! and V@ of MedleyDB
#medley_dir = ["/import/c4dm-datasets/MedleyDB_V1/V1", "/import/c4dm-datasets/MedleyDB_V2/V2"]
#paste the directory that contains the metadata files
meta_dir ="/homes/ssv02/Diff-MST/medley_metadata"



medley_split = {'train': [], 'test': [], 'val': []}
genre_test = []
genre_val = []
#load the metadata files
meta_files = glob.glob(os.path.join(meta_dir, "*.yaml"))
meta_files.sort()

for files in meta_files:
    
    file_name = os.path.basename(files).replace("_METADATA.yaml", "")
    print(file_name)
    
    
    with open(files, 'r') as f:
        data = load(f, Loader=Loader)
        if data['genre'] not in genre_test:
            medley_split['test'].append(file_name)
            genre_test.append(data['genre'])
        else:
            probability = np.random.rand()
            if probability < 0.8 and data['genre'] not in genre_val:
                medley_split['train'].append(file_name)
            else:
                medley_split['val'].append(file_name)

print(len(medley_split['train']))
print(len(medley_split['test']))
print(len(medley_split['val']))
```

```python
dump_dir = "/homes/ssv02/Diff-MST/"
dump_dir = os.path.join(dump_dir, "medley_split.yaml")
# if not os.path.exists(dump_dir):
#     os.makedirs(dump_dir)
with open (dump_dir, 'w') as f:
        dump(medley_split, f, Dumper=Dumper)
        print("dumped")
```
